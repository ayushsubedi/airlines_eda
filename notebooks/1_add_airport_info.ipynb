```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/Airline_Delay_Cause.csv')
```

```python
df_airports = pd.read_csv('../datasets/airports_our_world.csv')
```

```python
list(df_airports)
```

```python
df_airports = df_airports[['type', 'name',  'latitude_deg', 'longitude_deg','elevation_ft', 'iso_country', 'iso_region', 'municipality', 'gps_code', 'iata_code',]].iloc[1:].reset_index(drop=True)
```

```python
df_airports = df_airports[df_airports.iata_code.isin(df.airport)]
```

```python
df = df.set_index('airport').join(df_airports.set_index('iata_code')).reset_index().rename({'index': 'airport'}, axis=1)
```

```python
df.to_csv('../datasets/tableau.csv', index=False)
```
