```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
df = pd.read_csv('../datasets/tableau.csv')
```

```python
df
```

```python
df_airports = df[['airport', 'airport_name', 'municipality', 'iso_region', 'iso_country', 'elevation_ft','latitude_deg', 'longitude_deg']].drop_duplicates().dropna().reset_index(drop=True)
```

```python
df_airports.to_csv('airports_cleaned.csv', index=False)
```

```python
df = df[['airport',
 'year',
 'month',
 'carrier',
 'carrier_name',
 'arr_flights',
 'arr_del15',
 'carrier_ct',
 'weather_ct',
 'nas_ct',
 'security_ct',
 'late_aircraft_ct',
 'arr_cancelled',
 'arr_diverted',
 'arr_delay',
 'carrier_delay',
 'weather_delay',
 'nas_delay',
 'security_delay',
 'late_aircraft_delay']]
```

```python
df_airports
```

```python
conn = sqlite3.connect('../databases/airports_delay.db')
conn.execute("DROP TABLE IF EXISTS airports;")
conn.execute("DROP TABLE IF EXISTS delays;")
```

```python

with sqlite3.connect('../databases/airports_delay.db') as con:
    df_airports.to_sql('airports', con=con, index=False, dtype={'airport': 'TEXT PRIMARY KEY',
                                                     'airport_name': 'TEXT',
                                                     'municipality': 'TEXT',
                                                     'iso_region': 'TEXT',
                                                     'iso_country': 'TEXT',
                                                     'elevation_ft': 'REAL',
                                                     'latitude_deg':'REAL',
                                                     'latitude_deg': 'REAL'})
```

```python
pd.read_sql('select * from airports', conn)
```

```python
with sqlite3.connect('../databases/airports_delay.db') as con:
    df.to_sql('delays', con=con, index=False, dtype={'id': 'INTEGER PRIMARY KEY',
                                                     'airport': 'TEXT',
                                                     'year': 'TEXT',
                                                     'month': 'TEXT',
                                                     'carrier': 'TEXT',
                                                     'carrier_name': 'REAL',
                                                     'arr_flights':'REAL',
                                                     'arr_del15': 'REAL',
                                                     'carrier_ct': 'REAL',
                                                     'weather_ct': 'REAL',
                                                     'nas_ct': 'REAL',
                                                     'security_ct': 'REAL',
                                                     'late_aircraft_ct': 'REAL',
                                                     'arr_cancelled': 'REAL',
                                                     'arr_diverted': 'REAL',
                                                     'arr_delay': 'REAL',
                                                     'carrier_delay': 'REAL',
                                                     'weather_delay': 'REAL',
                                                     'nas_delay': 'REAL',
                                                     'security_delay': 'REAL',
                                                     'late_aircraft_delay': 'REAL'
                                                    })
```
